```python
def parseAuthor(author):
    names = author.split(",")
    # print(names)
    if len(names) > 2:
        # print(author)
        # print(names)
        return names[0] + names[1] + " "
    if len(names) == 2:
        # print(names[0] + ", " + names[1][1:2] + ".")
        return names[0] + ", " + names[1][1:2] + "."
    if len(names) == 1:
        print(author)
        return "ERROR"

def parseAuthors(authors):
    parsedAuthorString = ""
    if len(authors) <= 3:
        for author in authors:
            parsedAuthorString += parseAuthor(author)
    else:
        for author in authors[:3]:
            parsedAuthorString += parseAuthor(author)
        parsedAuthorString += "et al."
    return parsedAuthorString
```

```python
for authors in df['authors']:
    parseAuthors(authors)
```
